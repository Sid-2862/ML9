## Logistic Regression

# Read the dataset

```python
path = (
    r"https://raw.githubusercontent.com/Sindhura-tr/Datasets/refs/heads/main/Iris.csv"
)
import pandas as pd
df = pd.read_csv(path)
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

## Separate x and y features
y:species

```python
x = df.drop(columns=["Id","Species"])
y = df[["Species"]]
```

```python

x.head()
```

```python
y.head()
```

## Data Preprocessing and data cleaning

```python
# Data Preprocessing and Data Cleaning
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = make_pipeline(SimpleImputer(strategy="mean"),
                         StandardScaler()
                         ).set_output(transform="pandas")
```

```python
num_pipe
```

```python
x_pre = num_pipe.fit_transform(x)
x_pre
```

## Split the data into training and testing

```python
# Train Test Split
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(
    x_pre, y, train_size=0.85, random_state=42
)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

## Model Building

```python

from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(xtrain,ytrain)
```

```python

model.classes_
```

```python
model.coef_
```

```python

model.intercept_
```

## Model Evaluation

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest, ytest)
```

```python
ypreds = model.predict(xtest)
ypreds
```

```python
yprobs = model.predict_proba(xtest)
yprobs[:5]
```

## Confusion Matrix, Classification Report

```python
from sklearn.metrics import ConfusionMatrixDisplay,classification_report
ConfusionMatrixDisplay.from_estimator(model,xtest,ytest)
```

```python
print(classification_report(ytest,ypreds))
```

```python

from sklearn.metrics import f1_score
```

```python

f1_test = f1_score(ytest, ypreds, average="macro")
f1_test
```

## Check for cross validated scroes

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, xtrain, ytrain, cv=3, scoring="f1_macro")
scores
```

```python
scores.mean()
```

## From the above testing scores,f1_macro scores and cross validated scores, we can consider this model for final predictions

```python

path2 = r"https://raw.githubusercontent.com/Sindhura-tr/Datasets/refs/heads/main/Iris%20Sample.csv"
xnew = pd.read_csv(path2)
xnew.head()
```

```python
xnew = xnew.drop(columns=["Unnamed: 4"])
```

```python
xnew_pre = num_pipe.transform(xnew)
xnew_pre
```


## Out of sample predictions, saving of data to dataframe and saving the results to csv file

```python

final_preds = model.predict(xnew_pre)
final_preds
```

```python

xnew["Species Predicted"] = final_preds
xnew
```

```python
xnew.to_csv("Species Predicted.csv", index=False)
```
